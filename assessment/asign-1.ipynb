```python
# Create a Python application for below functionalities. 
# a. Items — to store and access the list of items. 
# b. Add functionality 
# c. Remove Functionality. 
# e. View All items. 
# Apply using Object Oriented Programming. 
# Handle the Exceptions appropriately. 
# Make it a Dynamic Application. 
```

```python
from abc import ABC, abstractmethod
```

```python
class Shopping(ABC):
    def __init__(self):
        self.items = []

    @abstractmethod
    def neitems(self):
        self.items = []

    @abstractmethod
    def add_items(self, item):
        try:
            self.items.append(item)
            print(f"Item '{item}' added successfully.")
        except Exception as e:
            print(f"Error occurred while adding item: {e}")

    @abstractmethod
    def del_items(self, item):
        try:
            if item in self.items:
                self.items.remove(item)
                print(f"Item '{item}' removed successfully.")
            else:
                print(f"Item '{item}' not found.")
        except Exception as e:
            print(f"Error occurred while removing item: {e}")

    @abstractmethod
    def view_items(self):
        try:
            if self.items:
                print("Items in the cart:")
                for item in self.items:
                    print(item)
            else:
                print("No items in the cart.")
        except Exception as e:
            print(f"Error occurred while viewing items: {e}")

class Shop(Shopping):
    def neitems(self):
        super().neitems()
    def add_items(self, item):
        super().add_items(item)
    def del_items(self, item):
        super().del_items(item)
    def view_items(self):
        super().view_items()

def default_fun():
    try:
        num_items = int(input("Enter the number of items to add: "))
        cart = Shop()
        for _ in range(num_items):
            item = input("Enter item to add: ")
            cart.add_items(item)
        cart.view_items()
        item_to_remove = input("Enter item to remove: ")
        cart.del_items(item_to_remove)
        cart.view_items()
    except ValueError as e:
        print(e)
    except Exception as e:
        print(e)
default_fun()
```

```python
from abc import ABC, abstractmethod

class BankAccount(ABC):
    accounts = {}

    def __init__(self, account_number, initial_balance=0):
        self.account_number = account_number
        self.balance = initial_balance
        BankAccount.accounts[self.account_number] = self.balance  

    @abstractmethod
    def deposit(self,amount):
        try:
            if amount<=0:
                raise ValueError("Deposit amount must be positive.")
            self.balance+=amount
            BankAccount.accounts[self.account_number] = self.balance  
            print(f"Deposit of {amount:.2f} successful.")
        except ValueError as e:
            print(e)

    @abstractmethod
    def withdraw(self,amount):
        try:
            if amount<=0:
                raise ValueError("Withdrawal amount must be positive.")
            if self.balance<amount:
                raise ValueError("Insufficient balance.")
            self.balance-=amount
            BankAccount.accounts[self.account_number] = self.balance 
            print(f"Withdrawal of {amount:.2f} successful.")
        except ValueError as e:
            print(e)

    @abstractmethod
    def available_balance(self):
        print(f"Available balance for {self.account_number}: {self.balance:.2f}")

    @abstractmethod
    def update_information(self,amount):
        if amount != 0:
            transaction_details = {
                'account_number': self.account_number,
                'amount': amount
            }
      
class acc(BankAccount):
    def deposit(self,amount):
        super().deposit(amount)

    def withdraw(self,amount):
        try:
            if self.balance == 0:
                raise ValueError("Cannot withdraw from account with zero balance.")
            super().withdraw(amount)
        except ValueError as e:
            print(e)

    def available_balance(self):
        super().available_balance()

    def update_information(self,amount):
        super().update_information(amount)

def def_fun():
    try:
        account_number = int(input("Enter account number: "))
        if account_number<0:
            raise ValueError("Account number must be positive.")

        initial_balance = float(input("Enter initial balance: "))
        if initial_balance < 0:
            raise ValueError("Initial balance must be non-negative.")

        account = acc(account_number, initial_balance)

        deposit_amount = float(input("Enter deposit amount: "))
        if deposit_amount < 0:
            raise ValueError("Deposit amount must be positive.")
        account.deposit(deposit_amount)

        withdraw_amount = float(input("Enter withdrawal amount: "))
        if withdraw_amount<0:
            raise ValueError("withdrawal amount cannot be negative")
        account.withdraw(withdraw_amount)
        
        account.available_balance()

        update_amount = float(input("Enter update amount: "))
        account.update_information(update_amount)

    except ValueError as e:
        print(e)

def_fun()
```
