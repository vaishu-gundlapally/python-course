```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_excel('customer.xlsx')
data
```

```python
data.drop_duplicates()
```

```python
df=data.drop(columns='Not_Useful_Column')
```

```python
df
```

```python
df["Phone_Number"]=df["Phone_Number"].str.replace('N/a','0')
```

```python
df["Phone_Number"]=df["Phone_Number"].fillna(0)
```

```python
df
```

```python
df[["street","state","zip"]] = df["Address"].str.split(',',n=2,expand=True)
df
```

```python
df1=df.drop(columns="Address")
df1
```

```python
df1["Paying Customer"]=df1["Paying Customer"].str.replace("N/a","0")
df1
```

```python
df1["Paying Customer"]=df1["Paying Customer"].apply(lambda x:"Yes" if x=='Y' else x)
```

```python
df1["Paying Customer"]=df1["Paying Customer"].apply(lambda x:"No" if x=='N' else x)
```

```python
df1
```

```python
df1["Do_Not_Contact"]=df1["Do_Not_Contact"].apply(lambda x:"Yes" if x=='Y' else x)
```

```python
df1["Do_Not_Contact"]=df1["Do_Not_Contact"].apply(lambda x:"No" if x=='N' else x)
df1
```
